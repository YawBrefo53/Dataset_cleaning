# Lets start the data cleaning process
 using the info function and the isnull.sum creates the impression that there are no missing values.
 sometimes there are placeholder values.
 so we will check for that in some of the columns.

```python
df.head(2)
```

```python
# we will use the value count function to check for the placeholder values
# we will do the data cleaning by doing it column by column
# after dealing with a column, we move on to the next column
# this is the format that we are going to use, which lets us focus on a column at a time

# we will start with the first column job title

df['Job_Title'].value_counts(dropna=False)
```

for the job title, there is no a placeholder value, which means no missing values as well. the datatype is also ok

```python
# the next column is the salary estimate column

df['Salary_Estimate'].value_counts(dropna=False)
```

There is also no placeholder in the salary estimate column. 
For this column will split it into minimum and maximum salary.
To do that, we will first remove the parenthesis and its content before splitting

```python
# using the replace function to remove the parenthesis with the content
# this is a simple understanding way to do it
# lets use regex

pattern = r'\s*\([^)]*\)'

df['Salary_Estimate'] = df['Salary_Estimate'].str.replace(pattern,'', regex=True)
df['Salary_Estimate'].head()
```

```python
# we strip the salary estimate column for any trailing spaces
df['Salary_Estimate'] = df['Salary_Estimate'].str.strip()
df.head(2)
```

```python
# the next task is to split the columns using the '-' as the delimeter
# to do that, will add the two columns and assign the split of the salary estimate to them

df[['Min_Salary', 'Max_Salary']] = df['Salary_Estimate'].str.split('-', expand=True)
df.head(2)
```

The two columns have been added with the values. we can now drop the estimate column

```python
df = df.drop('Salary_Estimate', axis=1)
df.head(2)
```

We are done with the task on the Salary estimate column which is now the min and max salary columns.


```python
# Our next colum is the rating column but we can see that Company name column also has the ratings attached to it
# to resolve this, lets extract the rating in the company column and then decide on the one to use.

df[['Company', 'Ratings']] = df['Company_Name'].str.split('\n', expand=True)
df.head(2)

```

Now that we have the Rating column and Ratings from the company name, we can decide on which one to use. I will use the value count functions on the two columns
